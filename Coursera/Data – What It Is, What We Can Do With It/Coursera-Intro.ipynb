## Theories
* Descriptive
    - conjecture about the characteristics of a phenomena
* Causal
    - conjecture about the causes of a phenomena
    requires dependent and independent variables. The independent variables causes the phenomena, the dependent variable is the result 
    - causal theories suggest that an observable outcome has a specific cause

### characteristics of a good theory
* positive (not normative)
    - a positive theory describes how the world is, whereas a normative theory describes how the world ought to be. The difficulty with normative theories from a scientific perspective is that they are impossible to prove correct or incorrect. In contrast a positive theory can be proven or disproven with empirical evidence
        example:
        normative: enhanced interrogation techniques are unjust and should be abolished
        postive: enhanced interrogation techniques produce false information
* observable implications
    - good theories predict that something observable will occur
    - to prove or disprove the theory it is enough to show that the predicted observable result did or did not occur
* testable
    - if a theory is unfalsifiable -- that is if everything we observe in the world is compatible with the theory -- it cannot be empirically tested
    - to be testable:
        1) necessary data must be available
        2) adequate time to test the theory is needed
        3) testing procedure must be ethical
* generalizable
    - applies to more than one case, that is it should apply to more than one instance of a phenomenon
* parsimonious
    - do not contain unneccesary parts that complicate the issue
    - simpler is generally better

## Theories in Scientific Research
> a scientific theory is a system of constructs (concepts) and propositions (relationships between those constructs) that collectively presents a logical, systematic, and coherent explanation of a phenomenon of interest within some assumptions and boundary conditions (Bacharach 1989)

Theories should explain why things happen, rather than being simply predictive. Prediction only needs correlation, wheras a theory should encompass causation. To establish causation, three things are needed:
1) correlation between two constructs
2) temporal precedence (the cause must preced the effect)
3) alternative explanations as to cause must be able to be disproven through testing

## Causal Effect and Potential Outcomes Framework
Causal questions have two outcomes, factual and counterfactual. Factual outcomes are what happens in the presence of a given manipulation, whereas counterfactual outcomes are what happens in the absence fo that manipulation
The true causal effect is the difference between the factual outcome and the counterfactual outcome. But both cannot be observed for the same population with the same inputs at the same time. Hence the counterfactual is by nature unobserved (and speculative?)
For a group of observations, you can divide inviduals into two groups.  a group of individuals subjected to an intervention and group of individuals not subjected to the intervention (the control group).
The control group can be used to estimate the counterfactual for the treatment units.
Because we can only observe one outcome for each person, and because there are many variables outside of the measurement scopt that can effect the outcomes, estimatin of causal effects is challenging
